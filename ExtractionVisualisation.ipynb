```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
def plotExtraction(filename, maxY=500, printTime=True, zoom=None):
    extraction_df = pd.read_csv("./onedatacustom/resultQuery/"+filename+".csv")
    extraction_df['diff']=extraction_df['Extractiontime']-extraction_df['ctime']
    extraction_df['relative_ingest_time']=extraction_df['ctime']-extraction_df['ctime'][0]
    if printTime:
        ax=extraction_df.plot(x='relative_ingest_time', y='diff', style ='o' , figsize=(17,15), title=filename)
        ax.set_xlabel("ingest time in 's'")
    else:
        ax=extraction_df.plot( y='diff', style ='o' , figsize=(17,15), title=filename)
        ax.set_xlabel("ID")
    
    ax.set_ylabel("extraction duration in 's'")
    ax.set_ylim(0,maxY)
  
    ax.set_title(filename)
    if zoom!= None:
        ax.set_xlim(zoom[0],zoom[1])
    print (extraction_df.shape)
```

```python
plotExtraction("extraction_0latency_500files_4directories_NoOptim",3500)
```
